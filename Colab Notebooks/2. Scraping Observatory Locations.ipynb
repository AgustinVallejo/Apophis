```python
continents = ['Africa',	'Antarctica',	'Asia',	'Australia',	'Europe',	'North America',	'South America',
              'Arctic Ocean',	'Atlantic Ocean', 'Indian Ocean',	'Pacific Ocean']
N = len(continents)
all_telescopes = pd.read_html("https://en.wikipedia.org/wiki/List_of_radio_telescopes")[:N]
```

```python
for c in range(N):
  continent = continents[c]
  telescopes = all_telescopes[c]
  telescopes['Longitude'] = None
  telescopes['Latitude'] = None
  for r,tel in telescopes.iterrows():
    try:
      lat, lon = wikipedia.WikipediaPage(wikipedia.search(tel.Name)[0]).coordinates
      telescopes.loc[r,'Longitude'] = float(lon)
      telescopes.loc[r,'Latitude']  = float(lat)
    except: pass
  all_telescopes[c] = telescopes
```

```python
writer = pd.ExcelWriter("/content/AllRadiotelescopes3.xlsx")

for c in range(N):
  continent = continents[c]
  telescopes = all_telescopes[c]
  telescopes.to_excel(writer,sheet_name=continent,)

writer.save()
```
